```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud


url = "https://raw.githubusercontent.com/JosephBARBIERDARNAL/data-matplotlib-journey/refs/heads/main/storms/storms.csv"
df_storms = pd.read_csv(url)

df_storms.head()
```

Your main objective is to create 5 unique charts using the provided datasets. 
We're providing a fullscreen sandbox here, but you can also work locally if you prefer!

### 1st plot: Pie chart

```python
df_storms_types = pd.DataFrame(df_storms.groupby('status')['n'].sum().reset_index())
labels = list(df_storms_types.status.unique())
n = list(df_storms_types.n)

fig, ax = plt.subplots()
ax.pie(n, labels=labels)
plt.show()
```

### 2nd Plot:  barplot

```python
fig, ax = plt.subplots()
ax.barh(labels, n)
plt.show()
```

### 3rd plot: Wordcloud plot

```python

freq = df_storms.groupby("status")["n"].sum().to_dict()

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='viridis'
).generate_from_frequencies(freq)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Frequency of storm types")
plt.show()
```

### 4th plot: donut plot

```python
freq = df_storms.groupby("status")["n"].sum()

plt.pie(
    freq.values,
    labels=freq.index,
    autopct="%1.1f%%",
    startangle=90,
    wedgeprops={"width": 0.4}
)

plt.title("Distribution of storm types")
plt.show()
```

## Que aprenderemos?

- Estructura de matplotlib (figure, axes)
- Graficos de lineas y puntos
- Barras y histogramas
- Personalizar titulos, ejes, leyendas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Estilo limpio
plt.style.use('seaborn-v0_8-whitegrid')
```

---

## 1. Primer grafico

```python
# Datos
months = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
trips = [1200, 1350, 1500, 1450, 1600, 1800]

# Grafico simple
plt.plot(months, trips)
plt.show()
```

### Anadir titulo y etiquetas

```python
plt.figure(figsize=(10, 6))
plt.plot(months, trips, marker='o')

plt.title("Viajes mensuales en bici", fontsize=14)
plt.xlabel("Mes")
plt.ylabel("Numero de viajes")

plt.show()
```

---

## 2. Figure y Axes (forma recomendada)

```python
# Crear figura y ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar en el eje
ax.plot(months, trips, marker='o', linewidth=2, markersize=8)

# Personalizar
ax.set_title("Viajes mensuales en bici", fontsize=14)
ax.set_xlabel("Mes")
ax.set_ylabel("Numero de viajes")

plt.show()
```

---

## 3. Multiples lineas

```python
# Datos por estacion
sol = [500, 550, 600, 580, 650, 720]
atocha = [400, 450, 500, 480, 520, 580]
retiro = [300, 350, 400, 390, 430, 500]

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(months, sol, marker='o', label='Sol')
ax.plot(months, atocha, marker='s', label='Atocha')
ax.plot(months, retiro, marker='^', label='Retiro')

ax.set_title("Viajes por estacion")
ax.set_xlabel("Mes")
ax.set_ylabel("Viajes")
ax.legend()

plt.show()
```

---

## 4. Grafico de barras

```python
stations = ["Sol", "Atocha", "Cibeles", "Retiro"]
total_trips = [3600, 2930, 2100, 2370]

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(stations, total_trips, color='steelblue')

ax.set_title("Total de viajes por estacion")
ax.set_xlabel("Estacion")
ax.set_ylabel("Total viajes")

plt.show()
```

### Barras horizontales

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.barh(stations, total_trips, color='steelblue')

ax.set_title("Total de viajes por estacion")
ax.set_xlabel("Total viajes")

plt.show()
```

### Barras agrupadas

```python
# Datos
subscribers = [2500, 2000, 1400, 1600]
casual = [1100, 930, 700, 770]

x = np.arange(len(stations))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(x - width/2, subscribers, width, label='Subscribers', color='steelblue')
ax.bar(x + width/2, casual, width, label='Casual', color='coral')

ax.set_title("Viajes por tipo de usuario")
ax.set_xticks(x)
ax.set_xticklabels(stations)
ax.legend()

plt.show()
```

---

## 5. Histograma

```python
# Generar duraciones
np.random.seed(42)
durations = np.random.exponential(scale=15, size=1000)

fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(durations, bins=30, color='steelblue', edgecolor='white')

ax.set_title("Distribucion de duraciones")
ax.set_xlabel("Duracion (min)")
ax.set_ylabel("Frecuencia")

plt.show()
```

### Histogramas superpuestos

```python
np.random.seed(42)
subscribers_dur = np.random.normal(loc=18, scale=5, size=500)
casual_dur = np.random.normal(loc=25, scale=8, size=500)

fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(subscribers_dur, bins=25, alpha=0.6, label='Subscribers', color='steelblue')
ax.hist(casual_dur, bins=25, alpha=0.6, label='Casual', color='coral')

ax.set_title("Duracion por tipo de usuario")
ax.set_xlabel("Duracion (min)")
ax.set_ylabel("Frecuencia")
ax.legend()

plt.show()
```

---

## 6. Scatter plot

```python
np.random.seed(42)
distances = np.random.uniform(1, 10, 100)
durations = distances * 3 + np.random.normal(0, 3, 100)

fig, ax = plt.subplots(figsize=(10, 6))

ax.scatter(distances, durations, alpha=0.6)

ax.set_title("Relacion distancia-duracion")
ax.set_xlabel("Distancia (km)")
ax.set_ylabel("Duracion (min)")

plt.show()
```

---

## 7. Subplots (multiples graficos)

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Grafico 1: Linea
axes[0, 0].plot(months, trips, marker='o')
axes[0, 0].set_title("Evolucion mensual")

# Grafico 2: Barras
axes[0, 1].bar(stations, total_trips, color='steelblue')
axes[0, 1].set_title("Total por estacion")

# Grafico 3: Histograma
axes[1, 0].hist(durations, bins=20, color='steelblue', edgecolor='white')
axes[1, 0].set_title("Distribucion duraciones")

# Grafico 4: Scatter
axes[1, 1].scatter(distances, durations, alpha=0.6)
axes[1, 1].set_title("Distancia vs Duracion")

plt.tight_layout()
plt.show()
```

---

## 8. Personalizar colores y estilos

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(months, sol, 
        color='#2ecc71',      # Color hex
        linewidth=2.5,         # Grosor linea
        linestyle='--',        # Estilo linea
        marker='o',            # Marcador
        markersize=10,         # Tamano marcador
        markerfacecolor='white',
        markeredgewidth=2,
        label='Sol')

ax.plot(months, atocha,
        color='#3498db',
        linewidth=2.5,
        linestyle='-',
        marker='s',
        markersize=8,
        label='Atocha')

ax.set_title("Viajes por estacion", fontsize=14, fontweight='bold')
ax.set_xlabel("Mes", fontsize=12)
ax.set_ylabel("Viajes", fontsize=12)
ax.legend(loc='upper left', frameon=True)

# Limites del eje Y
ax.set_ylim(0, 800)

plt.show()
```

---

## Resumen

| Grafico | Metodo |
|---------|--------|
| Linea | `ax.plot()` |
| Barras | `ax.bar()`, `ax.barh()` |
| Histograma | `ax.hist()` |
| Scatter | `ax.scatter()` |
| Subplots | `plt.subplots(rows, cols)` |

**Personalizacion:**
- `set_title()`, `set_xlabel()`, `set_ylabel()`
- `legend()`, `set_xlim()`, `set_ylim()`
- `figsize`, `color`, `linewidth`, `marker`
